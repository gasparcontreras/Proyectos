```python
import pandas as pd, datetime, os, time, pyautogui
```

```python
df = pd.read_excel('ALUMNOS PYTHON-J1.xlsx')
print(df.head())
```

```python
columnas_seleccionadas = ['Nombre y apellidos', 'Número de celular', 'Nota final']
df['Nota final'] = df['Nota final'].round(1)
df_seleccionado = df[columnas_seleccionadas]
print(df_seleccionado.head())
```

```python
condicion=(df['Nota final']!=0)
df=df[condicion]
print(df.shape)
```

```python
for index, row in df.iterrows():
    nombre = row['Nombre y apellidos']
    celular = row['Número de celular']
    nota_final = row['Nota final']
    
    mensaje = f"Hola {nombre}, tu nota final es: {nota_final}"
    
    try:
        os.startfile(f"whatsapp://send?phone=+51{celular}&text={mensaje}")
        time.sleep(10)
        
        pyautogui.hotkey('enter') # enviar mensaje
        time.sleep(5)
        
        print(f"Mensaje enviado a {nombre}")
        
    except Exception as e:
        print(f"Error al enviar a {nombre}: {e}")
        with open('logs_envios.txt', 'a') as f:
            f.write(f'{datetime.datetime.now()} - FALLO - +51{celular}, Error: {e}\n')

print('Concluido')
```
